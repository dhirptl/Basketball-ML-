# Train player bbox (YOLO11n) — async-friendly

Run this cell and continue other work while training. Ensure the kernel uses the project venv (`pip install -r requirements.txt`).

```python
from pathlib import Path
from ultralytics import YOLO
import torch

def resolve_device():
    if torch.backends.mps.is_available():
        return "mps"
    if torch.cuda.is_available():
        return "cuda"
    return "cpu"

ROOT = Path.cwd()
if not (ROOT / "Basketball Players").exists():
    ROOT = ROOT.parent
data_yaml = ROOT / "Basketball Players" / "dataset.yaml"
device = resolve_device()
print("device:", device, "data:", data_yaml)

model = YOLO("yolo11n.pt")
model.train(
    data=str(data_yaml),
    epochs=50,
    imgsz=640,
    batch=16,
    device=device,
    project=str(ROOT / "spatialTracker"),
    name="playerWeights",
)
```
